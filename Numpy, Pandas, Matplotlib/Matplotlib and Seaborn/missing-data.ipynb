```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
sales_data = pd.read_csv('./data_set.csv')
sales_data.head(10)
```

```python
sales_data.shape
```

```python
sales_data.isnull()
```

```python
sales_data.isna().sum()
```

```python
na_counts = sales_data.drop(['Date', 'Temperature', 'Fuel_Price'], axis=1).isna().sum()
print(na_counts)
na_counts.shape[0]
```

```python
sb.barplot(x=na_counts.index.values, y=na_counts)

plt.xticks(rotation=90) 

for i in range (na_counts.shape[0]):
    count = na_counts[i]
    
    plt.text(i, count+250, count, ha = 'center', va='top')
```
